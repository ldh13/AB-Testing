### Hypothesis test:

- Null hypothesis: there is no difference in the number of conversions in either group
- Alternative hypothesis: the members of group B are more likely to convert
- The alpha level will be 0.05
- Practical significance set to 
- With the samples that we have and alpha set to 0.05 the power

Since we don't have any data prior to the test I will use the control group conversion rate as the baseline and the minimum practical effect will be set to 1% of that baseline, with alpha, sample size and a minimum practical effect I will finally be able to calculate the power of the test

```python
# hypothesis test

# first I need the number conversions on each group
BM = user_data.group == 'A'
groupA_converted, groupB_converted = user_data.loc[BM].converted.sum(), user_data.loc[~BM].converted.sum()

# checking that the numbers add up
assert groupA_converted + groupB_converted == len(activ_ids)

# also need the number of customers in each group
n_groupA, n_groupB = user_data.loc[BM].shape[0], user_data.loc[~BM].shape[0]

# calculate test-statistic and p-value
z_stat, p_value = proportions_ztest(count= [groupB_converted, groupA_converted],
                                    nobs= [n_groupB, n_groupA],
                                    alternative='two-sided')

print(f'The results of the hyposthesis test are as follows:\nz-stat: {z_stat}\np-value: {p_value}')
```

*If the assumptions made in the hypothesis test are met then we can reject the null hypothesis and conclude that customers in group B are more likely to convert*

```python
# I will now manually compute the hypothesis test to make sure that the computations are correct

# it will be a z-test to compare two population proportions
BM = user_data.group == 'A'
conv_A, n_A = user_data.loc[BM].converted.sum(), user_data.loc[BM].shape[0]
conv_B, n_B = user_data.loc[~BM].converted.sum(), user_data.loc[~BM].shape[0]

# checking that the numbers add up to the correct values
assert conv_A + conv_B == len(activ_ids)
assert n_A + n_B == user_data.shape[0]

# calculating proportions
prop_A = conv_A / n_A
prop_B = conv_B / n_B

# performing the test
p_hat = (prop_A * n_A + prop_B * n_B) / (n_A + n_B)
denominator = np.sqrt(p_hat * (1 - p_hat) * (1/n_A + 1/n_B))
numerator = prop_B - prop_A
test_stat = numerator / denominator

print(f'z-stat: {test_stat}')
```

***Hypothesis testing results***

The results from the test with alpha set to 0.05 allow us to reject the Null hypothesis in favor on the alternative hypothesis with a confidence level of 95%.

This would indicate that customers that were in the treatment group had a higher conversion rate than those in the control group.

***Creating confidence intervals***

I will now build a confidence interval for the difference in proportions between the two groups.

```python
# confidence interval for the difference in the control and treatment groups

sample_statistic = prop_B - prop_A
standard_error = np.sqrt(
    prop_B * (1 - prop_B) / n_B + prop_A * (1 - prop_A) / n_A
)

z_value = 1.6448536269514722  # z value for a 95% CI (two-tailed) ?

# computing the CI
CI = sample_statistic - z_value * standard_error, sample_statistic + z_value * standard_error

CI
```

***Confidence interval results***

The 95% confidence interval of the difference in the proportion of the two groups is roughly (0.35%, 1.06%), although it seems a small difference the confidence interval does not contain 0 so we can conclude that the difference in proportions is statistically significant.

### Key takeaways
- After computing a hypothesis test and confidence interval we can conclude that the treatment group had a higher conversion rate than the control group

***Hypothesis testing***

- It would be useful to do more hypothesis tests to see whether these differences are significant and consider how that might affect the overall results.

- It would be important to do outlier analysis and see how these results would vary without said outliers.

```python
# hypothesis test for the difference in means
# alpha set to 0.05
# I will use a t-test
# right tailed test, group A has higher mean than group B

BM = activity_data_user.group == 'A'
mean_A, mean_B = activity_data_user.loc[BM].spent.mean(), activity_data_user.loc[~BM].spent.mean()
n_A, n_B = activity_data_user.loc[BM].count()[0], activity_data_user.loc[~BM].count()[0]
var_A, var_B = activity_data_user.loc[BM].spent.var(), activity_data_user.loc[~BM].spent.var()
pooled_var = ((n_A - 1) * var_A + (n_B - 1) * var_B) / (n_A + n_B - 2)
std_error = np.sqrt(pooled_var * (1/n_A + 1/n_B))
t_stat = (mean_A - mean_B) / std_error
t_stat
```

```python
# computing the hypothesis test using the stats library

t_stat, p_value = stats.ttest_ind(activity_data_user.loc[BM].spent,
                                  activity_data_user.loc[~BM].spent,
                                  alternative='greater')
print(f'The results of the test are as follows:\nT-stat: {t_stat}\nP-value: {p_value}')
```

***Outlier analysis***

- Compare the means from the two groups once the outliers have been removed
- Get the number of outliers on each group

```python
# I will manually find the outliers on each group using the iqr rule

BM = activity_data_user.group == 'A'
q1A, q3A = activity_data_user.loc[BM].spent.quantile(.25), activity_data_user.loc[BM].spent.quantile(.75)
q1B, q3B = activity_data_user.loc[~BM].spent.quantile(.25), activity_data_user.loc[~BM].spent.quantile(.75)
iqrA, iqrB = q3A - q1A, q3B - q1B
lower_capA, upper_capA = q1A - 1.5 * iqrA, q3A + 1.5 * iqrA
lower_capB, upper_capB = q1B - 1.5 * iqrB, q3B + 1.5 * iqrB

# now we subset for the data within the caps to remove the ouliers
BMA = (activity_data_user.loc[BM].spent > lower_capA) & (activity_data_user.loc[BM].spent < upper_capA)
mean_A_no_outliers = activity_data_user.loc[BM].loc[BMA].spent.mean()

BMB = (activity_data_user.loc[~BM].spent > lower_capB) & (activity_data_user.loc[~BM].spent < upper_capB)
mean_B_no_outliers = activity_data_user.loc[~BM].loc[BMB].spent.mean()

mean_A_no_outliers, mean_B_no_outliers
```

```python
# checking the number of outliers on each group

num_out_A = activity_data_user.loc[BM].id.nunique() - activity_data_user.loc[BM].loc[BMA].id.nunique()
num_out_B = activity_data_user.loc[~BM].id.nunique() - activity_data_user.loc[~BM].loc[BMB].id.nunique()

num_out_A, num_out_B
```

```python
# ok now I can perform the computations
# let's see the variance between the amount spent in either group

user_activity.groupby('group').spent.var()
```

```python
# performing a t-test with unequal variance to see if there is a difference in the mean amount spent across the two groups

BM = user_activity.group == 'A'
groupA, groupB = user_activity.loc[BM].spent, user_activity.loc[~BM].spent
t_stat, p_value = stats.ttest_ind(groupA, groupB, equal_var= False)
print(f'The result of the test are as follows:\nT-stat: {t_stat}\nP-value: {p_value}')
```

```python
# doing the computations

# calculate the mean and std of each group
meanA, meanB = groupA.mean(), groupB.mean()
stdA, stdB = groupA.std(ddof= 1), groupB.std(ddof= 1)

# calculate the standard error of the difference between the means
nA, nB = len(groupA), len(groupB)
sed = np.sqrt(stdA**2/nA + stdB**2/nB)

# calculate the t-statistic
t_stat = (meanA - meanB) / sed 

# calculate degrees of freedom
df = (stdA**2/nA + stdB**2/nB)**2 / ((stdA**2/nA)**2/(nA - 1) + (stdB**2/nB)**2/(nB - 1))

# calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df= df))

print('The results of the test are as follows:')
print(f'T-stat: {t_stat}\nP-value: {p_value}')
```
