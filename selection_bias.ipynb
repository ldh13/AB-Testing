### Selection Bias

The following is an analysis of the customers in each of the two groups (control and treatment) to ensure that there is no selection bias in the experiment

```python
# imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 400
import seaborn as sns
```

```python
# loading data

from data_connection import user_data
```

```python
# Looking at the first rows

user_data.head()
```

***Analysis Plan***

In order to ensure that the results from the test can be attributed to the banner we must ensure that the two groups only differ on their group placement, to do so we will check the other features and look for differences

```python
# First of all let's compare the countries of customers in each group

group_poss = user_data.group.unique()
group_dic = {'A':'Control', 'B':'Treatment'}

for i, group in enumerate(group_poss):
    BM = user_data.group == group
    plt.subplot(1, 2, i + 1)
    user_data.loc[BM].country.value_counts().plot.barh()
    plt.title(f'{group_dic[group]} group')

plt.tight_layout();
```

We can see from the plot above that both groups have almost identical figures when comparing the country feature

```python
# Comparing gender across the two groups

sns.catplot(x='gender', col='group', 
            kind='count', data= user_data);
```

Once again we can see that there doesn't seem to be any gender bias when selecting customers into either group

```python
# Compating visit device across groups

sns.catplot(x='visit_device', col='group',
            kind='count', data= user_data);
```

No difference between the control and the treatment group

```python
# Comparing join_dt for both groups

group_poss = user_data.group.unique()

for i, group in enumerate(group_poss):
    plt.subplot(1, 2, i + 1)
    BM = user_data.group == group
    user_data.loc[BM].join_dt.value_counts().sort_values().plot.barh()
    plt.title(f'{group_dic[group]} group')

plt.tight_layout();
```

Once again there isn't any difference between the customers on each group based on the visit date

### Results

After comparing the control and treatment group there doesn't seem to be any indicator of selection bias and we can assume that any differences found between the two groups would be caused by the experiment.
